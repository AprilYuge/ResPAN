```python
# Precision
```

```python
import numpy as np
```

```python
# respan
count = 0
for i in range(0,1015):
  for j in range(0,25):
    if (t1[i,j] == awgan_data[i,j]) and (t1[i,j]==True):
      count+=1
count/np.sum(awgan_data*1)
```

```python
# iMAP
count = 0
for i in range(0,1015):
  for j in range(0,25):
    if (t1[i,j] == imap_data[i,j]) and (t1[i,j]==True):
      count+=1
count/np.sum(imap_data*1)
```

```python
# Seurat
count = 0
for i in range(0,1015):
  for j in range(0,25):
    if (t1[i,j] == seurat_data[i,j]) and (t1[i,j]==True):
      count+=1
count/np.sum(seurat_data*1)
```

```python
# Recall
```

```python
import numpy as np
```

```python
# respan
count = 0
dom = 0
for i in range(0,1015):
  for j in range(0,25):
    if (t1[i,j] == awgan_data[i,j]) and (t1[i,j]==True):
      count+=1
    if (awgan_data[i,j] == False) and (t1[i,j]==True):
      dom += 1
count/(count+dom)
```

```python
# imap
count = 0
dom = 0
for i in range(0,1015):
  for j in range(0,25):
    if (t1[i,j] == imap_data[i,j]) and (t1[i,j]==True):
      count+=1
    if (imap_data[i,j] == False) and (t1[i,j]==True):
      dom += 1
count/(count+dom)
```

```python
# seurat
count = 0
dom = 0
for i in range(0,1015):
  for j in range(0,25):
    if (t1[i,j] == seurat_data[i,j]) and (t1[i,j]==True):
      count+=1
    if (seurat_data[i,j] == False) and (t1[i,j]==True):
      dom += 1
count/(count+dom)
```
